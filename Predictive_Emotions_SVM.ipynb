```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
df = pd.read_csv("/content/emotions.csv")
df.head()
```

```python
X = df['text']
y = df['label']


vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_vectorized = vectorizer.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Train an SVM classifier
svm_model = LinearSVC()
svm_model.fit(X_train, y_train)

# Make predictions and evaluate
y_pred = svm_model.predict(X_test)

# Results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# … (after you’ve trained svm_model and fitted vectorizer)

# Step 5: Read user input from the console
user_input = input("Enter a sentence to analyze its emotion: ")

# Wrap it in a list (vectorizer expects an iterable of documents)
new_text = [user_input]

# Vectorize and predict
new_vec = vectorizer.transform(new_text)
pred_label = svm_model.predict(new_vec)[0]

# Map to emotion name (make sure this matches your dataset)
label_map = {
    0: "sad",
    1: "joy",
    2: "anger",
    3: "fear",
    4: "neutral"
}

# Output
print(f"Predicted Emotion: {label_map.get(pred_label, 'Unknown')}")
```
